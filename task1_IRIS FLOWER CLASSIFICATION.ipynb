```python
import numpy as np       #linear algebra
import pandas as pd      #data processing
```

```python
df=pd.read_csv("iris.csv") # read dataset
```

```python
df #return dataframe
```

```python
#tuple of form(rows,column)
df.shape
```

```python
#dropping the column
df=df.drop(columns=["Id"])
```

```python
df.head() # return 5 entries from starting
```

```python
                                                         #categorical data in to numeric data
#assign"1-iris setosa" "2-iris-vesicolor""3-iris verginica"
df["Species"].replace({"Iris-setosa":1, "Iris-versicolor":2,"Iris-virginica":3},inplace=True)
```

```python
df #return dataframe
```

```python
#creating array
x=pd.DataFrame(df,columns=["SepalLengthCm","SepalWidthCm","PetalLengthCm"]).values
```

```python
x
```

```python
y=df.Species.values.reshape(-1,1)
```

```python
y
```

 #import necessary library

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

#train test split 

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
k=6
knclr=KNeighborsClassifier(k)
```

train modelusing knn

```python
knclr.fit(x_train,y_train)
```

```python
y_pred=knclr=knclr.predict(x_test)
```

```python
metrics.accuracy_score(y_test,y_pred)*100
```
